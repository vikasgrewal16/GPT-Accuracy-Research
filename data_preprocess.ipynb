```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter

# Download the NLTK stopwords dataset (if not already downloaded)
nltk.download('stopwords')
nltk.download('punkt')

```

```python
# Convert the "Comments" column to strings and remove NaN values
data['Comments'] = data['Comments'].astype(str)

# Combine all comments into a single text
all_comments = ' '.join(data['Comments'])

# Tokenize the text and filter out non-alphabetic tokens
tokens = word_tokenize(all_comments)

# Define a list of stopwords to be removed
stop_words = set(stopwords.words('english'))

# Filter out stopwords and words with a length of 1
filtered_tokens = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stop_words and len(word) > 1]

# Calculate word frequencies
word_freq = Counter(filtered_tokens)

# Create a DataFrame from the word frequencies
word_freq_df = pd.DataFrame.from_dict(word_freq, orient='index', columns=['Frequency']).reset_index()
word_freq_df = word_freq_df.rename(columns={'index': 'Word'})

# Sort the DataFrame by word frequency in descending order
word_freq_df = word_freq_df.sort_values(by='Frequency', ascending=False)

# Save the word frequencies to a CSV file
word_freq_csv_filename = "word_frequencies.csv"
word_freq_df.to_csv(word_freq_csv_filename, index=False)

print(f"Word frequencies saved to '{word_freq_csv_filename}'.")
```

```python
import matplotlib.pyplot as plt

# List of specific words to include
specific_words = ["problems", "better", "issues", "bug", "bad", "bing", "human", "waste", "improve", "improvement"]

# Filter the DataFrame to include only the specific words
filtered_word_freq_df = word_freq_df[word_freq_df['Word'].isin(specific_words)]

# Create the bar graph for the specific words
plt.figure(figsize=(10, 6))
plt.bar(filtered_word_freq_df['Word'], filtered_word_freq_df['Frequency'], alpha=0.7, color='blue')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Frequency Count of Specific Words')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the bar graph
plt.show()
```

```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize the Sentiment Intensity Analyzer
sentiment = SentimentIntensityAnalyzer(lexicon_file="vader_lexicon.txt")

# Read the CSV file containing comments
csv_filename = "/Shikhar/GPT-Accuracy-Research/youtube_gpt_video_data1.csv"  # Replace with your CSV file
df = pd.read_csv(csv_filename)

# Create lists to store sentiment scores and labels
compound_scores = []
sentiment_labels = []

# Analyze sentiment for each comment
for comment in df['Comments']:
    if isinstance(comment, str):
        ss = sentiment.polarity_scores(comment)
        
        # Classify sentiment based on the compound score
        if ss['pos'] >= 0.5:
            sentiment_label = "Positive"
        elif ss['neg'] >= 0.5:
            sentiment_label = "Negative"
        else:
            sentiment_label = "Neutral"
        
        # Append compound scores and sentiment labels to lists
        compound_scores.append(ss)
        sentiment_labels.append(sentiment_label)
    else:
        compound_scores.append(None)
        sentiment_labels.append(None)
# Add compound scores and sentiment labels to the DataFrame
df['Compound Score'] = compound_scores
df['Sentiment Label'] = sentiment_labels

# Save the updated DataFrame to a new CSV file
updated_csv_filename = "sentiment_analysis_results.csv"
df.to_csv(updated_csv_filename, index=False)

print(f"Sentiment analysis results saved to '{updated_csv_filename}'.")
  
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file with the appropriate encoding
csv_filename = "sentiment_analysis_results.csv"  # Replace with your CSV file
df = pd.read_csv(csv_filename, encoding='ISO-8859-1')

# Count the number of each sentiment label
sentiment_counts = df['Sentiment Label'].value_counts()

# Create a pie chart
labels = sentiment_counts.index
sizes = sentiment_counts.values
colors = ['lightgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0)  # To explode the 1st slice (Positive)

plt.figure(figsize=(6, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.

# Set the title
plt.title("Sentiment Analysis of Comments")

# Add a legend
plt.legend(labels, title="Sentiment")

# Show the pie chart
plt.show()
```
